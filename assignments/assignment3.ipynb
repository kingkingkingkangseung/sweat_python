## 1. 데이터 불러오기

```python
# 여기에 판다스를 불러오세요. 없으면 설치도 해야합니다.
import pandas as pd

# [문제1] villagers.csv 파일을 불러와 df라는 이름의 DataFrame으로 저장하세요.
df = pd.read_csv('data\\villagers.csv')
df
```

```python
# [문제2] 데이터의 상위 5개 행을 출력하세요.
print(df.head())
```

```python
# [문제3] 데이터의 하위 3개 행을 출력하세요.
print(df.tail(3))
```

## 2. 데이터 구조 확인

```python
# [문제4] df의 전체 정보를 출력하세요.
df.info()
```

```python
# [문제5] 수치형 데이터의 요약 통계를 출력하세요.
df.describe()
```

```python
# [문제6] 전체 열 이름을 리스트로 출력하세요.
list(df.columns)
```

## 3. 열 정보 및 컬럼 조작

```python
# [문제7] df의 열 이름 중 첫 번째 열의 이름을 출력하세요.
df.columns[0]
```

```python
# [문제8] df의 열 중 '성격' 열을 Series 형태로 추출하여 출력하세요.
df['성격']
```

```python
# [문제9] df에서 '이름', '자금' 두 열만 추출하여 출력하세요.
df[['이름', '자금']]
```

## 6. 조건 필터링

```python
# [문제10] 자금이 5000 이상인 주민들만 출력하세요.
df[df['자금'] >= 5000]
```

```python
# [문제11] 성격이 '먹보'이고, 자금이 6000 이상인 주민들만 출력하세요.
# 10번 문제를 잘 활용하시면 좋습니다 :)
df[(df['성격'] == '먹보') & (df['자금'] >= 6000)]
```
