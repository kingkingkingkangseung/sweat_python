## 🐣 중급 과제 (이해도 및 응용)

### 1. 다차원 배열 연습

```python
arr = np.array([[10, 20, 30], [40, 50, 60], [70, 80, 90]])

# (1) 두 번째 행
print("두 번째 행:", arr[1])

# (2) 두 번째 열
print("두 번째 열:", arr[:,1])

# (3) 50과 60 추출
print("50과 60 추출:", arr[1, [1,2]])
```

## 🧠 심화 과제 (심화 학습 유도)

### 체온 데이터 분석

```python
temps = np.array([36.6, 36.7, 37.2, 36.5, 36.9, 37.1, 36.8, 37.0, 36.4, 36.6])
print("평균 체온:", np.mean(temps))
print("표준편차:", np.std(temps))
print("최댓값:", np.max(temps))
print("최솟값:", np.min(temps))
```
